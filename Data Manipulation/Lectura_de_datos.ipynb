```python
ssma_1 = ssma[['Age', 'Gender', 'Avg_Daily_Usage_Hours', 'Most_Used_Platform']]
ssma_1
```

```python
avg = ssma_1[ssma_1['Avg_Daily_Usage_Hours']>4.0]
avg
```

```python
ssma_1 = ssma_1.rename(columns={"Hours":"Horas", "Gender":"Sexo",
                                "Platforma": "Plataforma", "Age":"Edad"})
```

```python
ssma_2 = ssma_1[ssma_1["Horas"] > 4.0]
ssma_2
```

```python
ssma_2.info()
```

```python
group_ssma_2 = ssma_2.groupby(by = "Sexo")["Horas"].max()
group_ssma_2
```

```python
platform_ssma_2 = ssma_2.groupby(by= "Plataforma")["Horas"].max()
platform_ssma_2
```

```python
ssma_2.groupby(by="Plataforma")['Horas'].count()
```

```python
ssma_3 = ssma_2.groupby("Plataforma")["Horas"].size()
ssma_4 = ssma_3.sort_values(ascending=False)
ssma_4
```
